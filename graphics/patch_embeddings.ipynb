```python
from pathlib import Path
from PIL import Image

image_path = Path("../image.png")
w, h = 224, 224
img = Image.open(image_path).resize((w, h)).convert("RGB")
img
```

```python
import numpy as np

patch_size = 16
patches_w = w // patch_size
patches_h = h // patch_size

img_data = np.array(img)
print(img_data.shape)
img_data = img_data.reshape((patches_h, patch_size, patches_w, patch_size, 3))
img_data = np.transpose(img_data, (0, 2, 1, 3, 4))
img_data.shape
```

```python
img_rows = []
gap_size = 3
gap_color = np.array([1, 1, 1]) * 255
for row in range(img_data.shape[0]):
    img_cols = []
    for col in range(img_data.shape[1]):
        if img_cols:
            img_cols.append(np.ones((patch_size, gap_size, 3)) * gap_color)
        img_cols.append(img_data[row, col])
    if img_rows:
        img_rows.append(np.ones((gap_size, img_rows[-1].shape[1], 3))* gap_color)
        
    img_rows.append(np.concatenate(img_cols, axis=1))

new_img_data = np.concatenate(img_rows, axis=0)

new_img = Image.fromarray(new_img_data.astype(np.uint8))
new_img
```
