```python
from pathlib import Path
from PIL import Image

start_image = Image.open(Path("../local-assets", "recon_image.png"))
end_image = Image.open(Path("../local-assets", "edit_image.png"))

start_image
```

```python
end_image
```

```python
assert start_image.size == end_image.size, "Images must be the same size for interpolation"
```

```python
import numpy as np
from IPython.display import Image as IPImage

alphas = np.linspace(0, 1, num=20)  # 20 steps from start to end

images = [Image.blend(start_image, end_image, alpha=alpha) for alpha in alphas]
```

```python
images[0].save(
    "../local-assets/image-blend.gif",
    save_all=True,
    append_images=images[1:],  # Remaining frames
    duration=75,              # Display time per frame (ms)
    loop=0                     # 0 means infinite loop
)

IPImage(filename="../local-assets/image-blend.gif")
```
